# Análise de dados do Campeonato Brasileiro

### Este é um projeto de simples de análise de dados que tem como objetivo procurar e trazer à luz algumas estatísticas sobre o campeonato brasileiro de futebol.


#### Primeiramente, temos que entender quais dados nós temos disponíveis:
##### Temos dois datasets, sendo eles:
> Um dataset contendo dados sobre as partidas, nele temos:

|Informação| Descrição |
|-----|-----|
|Rodada|Rodada em que ocorreu a partida|
|Data|Data da partida|
|Hora|Hora em que se iniciou a partida|
|Dia|Dia da semana em que ocorreu a partida|
|Mandante| Time que é o mandante da partida|
|Visitante| Time que é o visitante da partida|
|Vencedor|Vencedor da partida(" - " significa empate)
|Arena| Estádio onde ocorreu a partida|
|Mandante Placar| Gols feitos pelo mandante|
|Visitante Placar|Gols feitos pelo visitante|
|Mandante Estado|Estado de origem do time mandante|
|Visitante Estado|Estado de origem do time visitante|
|Estado Vencedor|Estado do time que venceu a partida (" - " significa empate)|
